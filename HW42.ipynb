```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import datasets 
from sklearn import preprocessing
```

```python
# Dataset loading

d = datasets.load_wine()
X_array = np.array(pd.DataFrame(data=d.data))
print(X_array)
```

```python
# Data normalization 

d_norm = preprocessing.normalize(X_array, norm='max')
print(d_norm)
```

```python
# K-means 

from sklearn.cluster import KMeans

def kmean_cluster_analysis(d_norm, n):
    model_kmean = KMeans(n_clusters = n)
    model_kmean.fit(d_norm)

    prediction = model_kmean.predict(d_norm)
    print(prediction, type(prediction))

    cluster0 = d_norm[prediction == 0]
    cluster1 = d_norm[prediction == 1]
    cluster2 = d_norm[prediction == 2]
    cluster3 = d_norm[prediction == 3]
    cluster4 = d_norm[prediction == 4]
    
    plt.scatter(cluster0[:,0], cluster0[:,1], color = 'red')
    plt.scatter(cluster1[:,0], cluster1[:,1], color = 'black')
    plt.scatter(cluster2[:,0], cluster2[:,1], color = 'blue')
    plt.scatter(cluster3[:,0], cluster3[:,1], color = 'yellow')
    plt.scatter(cluster4[:,0], cluster4[:,1], color = 'green')
    
    plt.title(f'K = {n}')
    plt.show()
    
for i in range(2, 6):    
    kmean_cluster_analysis(d_norm, i)
```

```python
# DBSCAN  

from sklearn.cluster import DBSCAN
from sklearn.decomposition import PCA


def dbscan_cluster_analysis(d_norm, e, ms):

    model_dbscan = DBSCAN(eps=e, min_samples=ms)
    model_dbscan.fit(d_norm)
    print('model_dbscan.labels_\n', model_dbscan.labels_)

    pca = PCA(n_components=2).fit(d_norm)
    pca_2d = pca.transform(d_norm)
    pca_2d.shape

    for i in range(0, pca_2d.shape[0]):
        if model_dbscan.labels_[i] == 0:
            plt.scatter(pca_2d[i, 0], pca_2d[i, 1], c = 'r', marker = 'o')
        elif model_dbscan.labels_[i] == 1:
            c2 = plt.scatter(pca_2d[i, 0], pca_2d[i, 1], c = 'g', marker = '+')
        elif model_dbscan.labels_[i] == 2:
            c3 = plt.scatter(pca_2d[i, 0], pca_2d[i, 1], c = 'b', marker = 'x')
        elif model_dbscan.labels_[i] == -1:
            c4 = plt.scatter(pca_2d[i, 0], pca_2d[i, 1], c = 'y', marker = '>')  
    plt.title(f'eps = {e} min_samples = {ms}')
    plt.show()
            
dbscan_cluster_analysis(d_norm, 0.1, 5)
dbscan_cluster_analysis(d_norm, 0.01, 7)
dbscan_cluster_analysis(d_norm, 0.01, 15)
dbscan_cluster_analysis(d_norm, 0.01, 20)
dbscan_cluster_analysis(d_norm, 0.001, 20)
```

```python
# Hierarchical clustering

from scipy.cluster.hierarchy import linkage, dendrogram

def hierarchy_cluster_analysis(d_norm, method):
    mergings = linkage(d_norm, method = method)
    clusters = dendrogram(mergings, leaf_rotation=90)
    plt.title(method)
    plt.show()


methods = ['single', 'complete', 'average', 'weighted', 'centroid', 'median', 'ward']
for method in methods:
    hierarchy_cluster_analysis(d_norm, method)    
```
